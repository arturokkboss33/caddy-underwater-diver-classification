```python
from random import randrange
#from IPython.display import Image
from matplotlib.pyplot import imshow
%matplotlib inline

show_images = False
debug       = False
early_stop  = False

#dataset_dir = Path('/mnt/btrfs-data/dataset/caddy-gestures/CADDY_gestures_complete_v2_release/caddy-gestures-complete-v2-release-all-scenarios-fast.ai')
dataset_dir = Path('/mnt/btrfs-data/dataset/caddy-gestures/CADDY_gestures_complete_v2_release/all-scenarios-orig')

numbers_in_words = [ 'one', 'two', 'three', 'four', 'five' ]

fn = sentences_dir / 'caddian-1000-sentences-test-set.txt'

output_image_seq = []

sentences_counter = 0
with (open(fn, "r")) as openfile:
    line = openfile.readline()
    #files_dict = pickle.load(openfile)
    while line:
        #print(line)
        tokens = line.split()
        if len(tokens) == 0 or tokens[0] == 'Creating':
            line = openfile.readline()
            continue
        print(tokens)
        
        real_fn_list = []
        
        for tok in tokens:
            gst = tok
            for i in range(5):
                if gst == str(i):
                    gst = numbers_in_words[i-1]
            gst_len = len(files_dict['overall'][gst])
            idx = randrange(0, gst_len)
            #img_name = str(files_dict['overall'][gst][idx])
            img_name = files_dict['overall'][gst][idx]
            #print(img_name)
            #img_name.replace('all-scenarios-split-cristal-clear-vs-cloudy-and-dark/all-scenarios/' + gst + '/crystal-clear/', '')
            if debug:
                print(img_name)
                print(img_name.stem)
            real_fn = dataset_dir / img_name
            if debug:
                print(real_fn)
            if real_fn.is_file():
                real_fn_list.append(real_fn)
                if debug:
                    print('exists')
                    print('')
        
        
        output_image_seq.append(real_fn_list)
        
        line = openfile.readline()
        sentences_counter += 1

        if early_stop and sentences_counter >= 10:
            break
            
        if show_images and sentences_counter < 10 and len(tokens) < 10:
            size = 256,256
            #images = [Image.open(x).thumbnail(size, Image.ANTIALIAS) for x in real_fn_list]
            images = [Image.open(x).resize(size, Image.ANTIALIAS) for x in real_fn_list]
            
            widths, heights = zip(*(i.size for i in images))

            total_width = sum(widths)
            max_height  = max(heights)

            new_im = Image.new('RGB', (total_width, max_height))
            
            x_offset = 0
            for im in images:
                  new_im.paste(im, (x_offset,0))
                  x_offset += im.size[0]

            #imshow(np.asarray(new_im))
            new_im.show()
                
            '''
            for real_fn in real_fn_list:
                im = Image.open(real_fn)
                #im.show()
                imshow(np.asarray(im))
                
                #im = Image(filename=str(real_fn))
                #display(im)
            '''
            
print(f'Produced: {len(output_image_seq)} image sequences')
print(len(output_image_seq[7]), output_image_seq[0])

output_image_seq_fn = sentences_dir / 'output-image-sequences.pkl'
with open(str(output_image_seq_fn), 'wb') as fp:
    pickle.dump(output_image_seq, fp)
```
