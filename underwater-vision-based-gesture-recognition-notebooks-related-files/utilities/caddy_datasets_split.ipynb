```
with open(loadpath / 'train_valid-df-model-A.pkl', 'rb') as pickle_file:
    train_valid_df_model_a = pickle.load(pickle_file)
with open(loadpath / 'train_valid-df-model-B.pkl', 'rb') as pickle_file:
    train_valid_df_model_b = pickle.load(pickle_file)
with open(loadpath / 'train_valid-df-model-C.pkl', 'rb') as pickle_file:
    train_valid_df_model_c = pickle.load(pickle_file)
with open(loadpath / 'train_valid-df-model-F.pkl', 'rb') as pickle_file:
    train_valid_df_model_f = pickle.load(pickle_file)
    
classes = ['backwards', 'boat', 'carry', 'down', 'end_comm', 'five', 'four', 'here', 'mosaic', 'num_delimiter', 'one', 'photo', 'start_comm', 'three', 'two', 'up']

print(30*'-')
print('Model A - train/valid sets')
print(30*'-')
for cl in classes:
    train_valid_len = len(train_valid_df_model_a[cl])
    train_len = int(train_valid_len*.8)
    valid_len = int(train_valid_len*.2)
    clj = cl.rjust(13, ' ')
    print(f'{clj} - {train_len} - {valid_len}')
print(30*'-')


print(30*'-')
print('Model B - train/valid sets')
print(30*'-')
for cl in classes:
    train_valid_len = len(train_valid_df_model_b[cl])
    train_len = int(train_valid_len*.8)
    valid_len = int(train_valid_len*.2)
    clj = cl.rjust(13, ' ')
    print(f'{clj} - {train_len} - {valid_len}')
print(30*'-')


print(30*'-')
print('Model C - train/valid sets')
print(30*'-')
for cl in classes:
    train_valid_len = len(train_valid_df_model_c[cl])
    train_len = int(train_valid_len*.8)
    valid_len = int(train_valid_len*.2)
    clj = cl.rjust(13, ' ')
    print(f'{clj} - {train_len} - {valid_len}')
print(30*'-')



print(30*'-')
print('Model F - train/valid sets')
print(30*'-')
for cl in classes:
    train_valid_len = len(train_valid_df_model_f[cl])
    train_len = int(train_valid_len*.8)
    valid_len = int(train_valid_len*.2)
    clj = cl.rjust(13, ' ')
    print(f'{clj} - {train_len} - {valid_len}')
print(30*'-')
```
